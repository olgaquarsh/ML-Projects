```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
```

```python
parkinson_data = pd.read_csv('parkinson_data.csv')
parkinson_data
```

```python
y = parkinson_data['RPDE'] # target
```

```python
X = parkinson_data.drop(['RPDE', 'name'], axis =1)
X
```

```python
parkinson_data.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 494)
```

```python
y_train
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```
